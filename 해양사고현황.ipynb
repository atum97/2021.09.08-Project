### 해양사고 발생지역 전처리
#### (더미화, 넓은범위인 서해, 남해, 동해만 사용)

```python
df = df[df['해양사고발생지역(통계용)'].str.contains('남해영해|서해영해|동해영해')]
df.head(2)
```

```python
df['해양사고발생지역(통계용)'].value_counts()
```

```python
location_dummies = pd.get_dummies(df['해양사고발생지역(통계용)'])

location_dummies.columns = ['남해영해', '서해영해', '동해영해']

df.drop(columns='해양사고발생지역(통계용)', axis=1, inplace=True)

df = df.join(location_dummies)
```

```python
df.head()
```

### 계절 더미화

```python
df['계절'].value_counts()
```

```python
season_dummies = pd.get_dummies(df['계절'])

season_dummies.columns = ['봄', '여름', '가을', '겨울']

df.drop(columns='계절', axis=1, inplace=True)

df = df.join(season_dummies)
df.head()
```
